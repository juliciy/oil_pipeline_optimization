# 拟合工频泵power=f(flow)

```python
def predict_fixed_frequency_power(flow):
    return 395.6312 + 0.7192 * flow -5.2654 * pow(10,-5) * pow(flow, 2)
```

```python
gongpin_list= ['日仪增输日照站1P-5泵有功功率','日仪增输日照站1P-6泵有功功率'
              ,'日仪增输东海站2P-1电机功率','日仪增输东海站2P-2电机功率'
              ,'日仪增输观音站4P-1电机功率','日仪增输观音站4P-2电机功率'
              ,'日仪增输淮安站3P-5电机功率','日仪增输淮安站3P-6电机功率']
```

```python
time_delta = 60*2  # 时间差设置
```

```python
fit_func = []
for k in range(8):
#    k=0
    sql = """
    select tagv_desc,tagv_value,tagv_fresh_time from  fz_tag_view
    where tagv_desc in ('日照站出站流量检测','{}') AND tagv_fresh_time LIKE '2023-0%'
    ORDER BY tagv_fresh_time
    """.format(gongpin_list[k])
    conn = get_conn()
    result = {}
    with conn.cursor() as cursor:
        cursor.execute(sql)
        data = cursor.fetchall()

    source_data = pd.DataFrame(data)
    source_data['time_stamp'] = source_data['tagv_fresh_time'].astype('str').apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S').timestamp())



    columns = [gongpin_list[k]]
    filter_data = pd.DataFrame(columns=[gongpin_list[k],'日照站出站流量检测'])



    j=0
    for i in source_data[source_data['tagv_desc']=='日照站出站流量检测'].index[1:-1]: 
        filter_data.loc[j,'日照站出站流量检测'] = source_data.loc[i,'tagv_value']
        for u in range(-2,3): 
            if (source_data.loc[i+u,'tagv_desc'] in columns and abs(source_data.loc[i,'time_stamp']-source_data.loc[i+u,'time_stamp'])<time_delta):
                filter_data.loc[j,source_data.loc[i+u,'tagv_desc']] =  source_data.loc[i+u,'tagv_value']
        j+=1

    filter_data = filter_data[filter_data[gongpin_list[k]]>5]
    filter_data.index=range(len(filter_data))
#     filter_data

    print('该泵拟合数据量为：',len(filter_data))
    if len(filter_data)>10:
        filter_data2 = filter_data[(filter_data['日照站出站流量检测'] > 3500) & (filter_data[gongpin_list[k]] > 100)]

        filter_data3 = filter_data2.loc[:,['日照站出站流量检测',gongpin_list[k]]]

        if len(filter_data3)>1:
            print('开始拟合%s'%gongpin_list[k][4:12])

            # 提取自变量和因变量
            x = filter_data3['日照站出站流量检测'].astype("float")
            y = filter_data3[gongpin_list[k]].astype('float')

            # 使用polyfit进行拟合
            coefficients = np.polyfit(x, y, 2)
            a = coefficients[0]
            b = coefficients[1]
            c = coefficients[2]

            # 构建拟合方程
            fit_equation = f'y = {a:.2f} * x^2 + {b:.2f} * x + {c:.2f}'
            # 绘制数据和拟合曲线
            x_range = np.linspace(min(x), max(x), 100)
            y_fit = a * x_range ** 2 + b * x_range + c
            # 打印拟合的函数表达式
            fit_equation = f'y = {a:.12f} * x^2 + {b:.10f} * x + {c:.10f}'
            print('拟合函数为：',fit_equation)
            fit_func.append(fit_equation)

            # 画图检查
            plt.figure(figsize=(6,3))
            flow = np.linspace(3000, 5000, 100) 
            yy = predict_fixed_frequency_power(flow)
            plt.plot(flow, yy, label='原始拟合')  # 原始拟合线
            plt.scatter(x, y, label='流量-power数据')  # 散点
            plt.plot(x_range, y_fit, color='red', label='当前拟合')  # 当前拟合线

            plt.xlabel('流量')
            plt.ylabel('功率')
            #plt.title('Quadratic Fit')
            plt.legend()
            plt.show()
        else:
            print('压差全部小于0.3','，此段时间泵未开启')
    else:
        print('该泵数据太少，无法拟合。')
```

```python
# 原始拟合函数
def power_function(freq, flow_capacity):
    return (((freq+0.6)**3)*395.6312 + ((freq+0.6)**2)*0.7192*flow_capacity +
            (freq+0.6)*(-5.2654)*1e-5*np.power(flow_capacity,2)) * 1.05
# 生成数据范围
freq_range = np.linspace(0.2, 0.4, 100)
flow_range = np.linspace(3000, 5000, 100)
freq_grid, flow_grid = np.meshgrid(freq_range, flow_range)
power_values = power_function(freq_grid, flow_grid)

# 定义拟合方程
def target_function(xy, a, b, c):
    x, y = xy
    return a*(x+0.6)**2 + b*(x+0.6)*y + c*y**2
```

```python
# 生成示例数据
x = (pump5['变频器1']-30)/50
y = pump5['流量']
z = pump5['变频5功率']

# 进行拟合
initial_guess = [1, 1, 1]  # 初始猜测值
params, covariance = curve_fit(target_function, (x, y), z, p0=initial_guess)
# 提取拟合后的参数
a_fit, b_fit, c_fit = params

# 生成拟合后的数据
x_grid, y_grid = np.meshgrid(np.linspace(min(x), max(x), 100), np.linspace(min(y), max(y), 100))
z_fit = target_function((x_grid, y_grid), a_fit, b_fit, c_fit)

# 绘制原始数据和拟合曲面
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter( y,x, z, c='b', label='Data')
ax.plot_surface( y_grid,  x_grid, z_fit, color='r', alpha=0.5, label='Fitted Surface')
ax.plot_surface(   flow_grid,     freq_grid,       power_values,     cmap='viridis')

ax.set_xlabel('flow')
ax.set_ylabel('freq')
ax.set_zlabel('power')
ax.set_title('Fitted Surface')
# plt.legend()
plt.show()
# 打印拟合的函数表达式
fit_equation = f'Fitted Equation: z = {a_fit:.8f} * (x + 0.6)**2 + {b_fit:.8f} * (x + 0.6) * y + {c_fit:.8f} * y**2'
print(fit_equation)
```
