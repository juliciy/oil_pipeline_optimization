```python
import pandas as pd
import numpy as np

from base.mysql_conn import get_conn
conn = get_conn()
sql = """
    select tagv_desc,tagv_value,tagv_fresh_time from  fz_tag_view
where tagv_desc in ('日照站出站流量检测','日仪增输日照站1P-6泵入口压力检测','日仪增输日照站1P-6泵出口压力检测') AND tagv_fresh_time LIKE '2023-07%'
ORDER BY tagv_fresh_time
    """
# print('get_started_num_bumps',sql)
result = {}

with conn.cursor() as cursor:
    cursor.execute(sql)
    data = cursor.fetchall()

source_data = pd.DataFrame(data)
source_data['time_stamp'] = source_data['tagv_fresh_time'].astype('str').apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S').timestamp())
source_data
```

```python
source_data['tagv_desc'].value_counts()
```

```python
columns = ['in_pressure', 'out_pressure','flow']
filter_Data = pd.DataFrame(columns=columns)

for i in range(len(source_data)//3):
#     print(abs(source_data.loc[i*3,'time_stamp']-source_data.loc[i*3+1,'time_stamp']))
#     print(abs(source_data.loc[i*3+1,'time_stamp']-source_data.loc[i*3+2,'time_stamp']))
    # print(source_data.loc[i*3+2,'time_stamp'])
    
    if (abs(source_data.loc[i*3,'time_stamp']-source_data.loc[i*3+1,'time_stamp'])<60 and abs(source_data.loc[i*3+1,'time_stamp']-source_data.loc[i*3+2,'time_stamp'])<60):
        filter_Data.loc[i,'in_pressure'] = source_data.loc[i*3,'tagv_value']
        filter_Data.loc[i,'out_pressure'] = source_data.loc[i*3+1,'tagv_value']
        filter_Data.loc[i,'flow'] = source_data.loc[i*3+2,'tagv_value']

filter_Data['pressure_diff'] = filter_Data['out_pressure'] - filter_Data['in_pressure']
filter_Data
```

```python
filter_Data2 = filter_Data[ (filter_Data['flow'] > 4600) & (filter_Data['pressure_diff'] > 0.3)]
if len(filter_Data2)>1:
    def do_predict(flow):
        return 1.6359 + 1.0057 * pow(10,-4) * flow + -2.0706 * pow(10,-8) * pow(flow, 2)
    flow = np.linspace(4650, 5000, 100) 
    y = do_predict(flow)
    
    plt.figure(figsize=(5,3))
    plt.plot(flow, y, label='y = 1.6359 + 1.0057e-4 * x - 2.0706e-8 * x^2')
    plt.scatter(filter_Data2['flow'],filter_Data2['pressure_diff'])
    plt.xlabel('flow',fontsize=12)
    plt.ylabel('pressure_diff',fontsize=12)
    #plt.title(‘电影时长与评分散点图’,fontsize=20)
    plt.show()
else:
    print(source_data.iloc[0,0][0:12],'，压差全部小于0.3','，此段时间泵未开启')
```
